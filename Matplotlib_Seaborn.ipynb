```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
df2=pd.DataFrame(np.random.randint(1,100,(4,3)),index=["harsh","rolli","golu","prinshu"],columns=["DSA","Php",".net"])
df2
```

```python
x =df2.index
y =df2["Php"]
plt.subplot(1,2,1)
plt.plot(x, y,"g^-")
plt.xlabel("marks")
plt.ylabel("students")
plt.title("php marks ")
plt.grid(color='b', linestyle='--' )
plt.ylim(0,100)
plt.legend(['First line', 'Second line'])
plt.show()




x =df2.index
y =df2["DSA"]
plt.subplot(1,2,1)
plt.plot(x, y,"r", marker='*', linestyle='-',data=df2)
plt.xlabel("marks")
plt.ylabel("students")
plt.title("DSA marks ")
```

```python
x =df2.index
y =df2["Php"]
plt.plot(x, y,"g^-")
plt.xlabel("student name")
plt.ylabel("student marks")
plt.title(" marks ")
y =df2["DSA"]
plt.plot(x, y,"r*-")
plt.grid(color='r', linestyle='--' )
plt.ylim(0,100)
y =df2[".net"]
plt.plot(x, y,"y*-")

plt.legend(['Php','DSA',".net"])
```

```python
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax2=fig.add_axes([1.1,1.2,0.5,0.5])


ax2.plot(x,y,'b')
ax2.set_xlabel("student name")
ax2.set_ylabel("student marks")
ax2.set_title("stu")
ax.plot(x,y,'b')
ax.set_xlabel("student name")
ax.set_ylabel("student marks")
ax.set_title("stu")
```
